## 3.  Fetching Data Using Web Scraping

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
import requests
url='https://www.gharbazar.com/property/list/normal/?_vt=1'
headers = {
    'User-Agent': 'Mozilla/15.0 (Macintosh; Intel Mac OS 10.14; rv:66.0) Gecko/20100101 Firefox/66.0'
}

webpage=requests.get(url,headers=headers).text
# response = requests.get(url, headers=headers)

soup=BeautifulSoup(webpage,'lxml')
```

```python
a=soup.find('div',class_='dropdown hide').text
```

```python
print(a.strip())
```
