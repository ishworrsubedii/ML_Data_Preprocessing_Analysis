# 2) Linear Regression

```python
salary_dataset=pd.read_csv('Salary_Dataset.csv')
X_train=salary_dataset.YearsExperience.values.reshape(-1,1)
Y_train=salary_dataset.Salary.values.reshape(-1,1)
```

```python
salary_dataset
```

```python
plt.scatter(X_train,Y_train)
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)
y_hat=lr.predict(X_train)
```

```python
plt.scatter(X_train,Y_train,marker='x',c='r',label='Actual Data')
plt.plot(X_train,y_hat,label='Model Prediction')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()
plt.show()

```

# 3) Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
```

```python
degree=1
poly=PolynomialFeatures(degree,include_bias=False)
x_train_norm_poly=poly.fit_transform(X_train)
x_train_norm_poly
```

```python
norm=StandardScaler()
```

```python
x_poly_norm=norm.fit_transform(x_train_norm_poly)
```

```python
plt.scatter(x_poly_norm,Y_train,marker='x',c='r',label='Actual Data')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()
plt.show()

```

```python
lr_norm=LinearRegression()
lr_norm.fit(x_poly_norm,Y_train)
model_pred_norm=lr_norm.predict(x_poly_norm)
```

```python
plt.scatter(x_poly_norm,Y_train,marker='x',c='r',label='Actual Data')
plt.plot(x_poly_norm,model_pred_norm,label='Model Prediction')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()
plt.show()

```

#### Normal vs Feature Scaled model mean squared error

```python
from sklearn.metrics import mean_squared_error
normalized_data_mse=mean_squared_error(Y_train,model_pred_norm)
```

```python
normalized_data_mse
```

```python
without_normalization=mean_squared_error(Y_train,y_hat)
```

```python
without_normalization
```

# 3) Model Evaluation

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_poly_norm,Y_train,test_size=0.1,random_state=0)
```

```python
print(f'''
x_train:{x_train.shape}
y_train:{y_train.shape}
x_test:{x_test.shape}
y_test:{y_test.shape}




''')
```

```python
linear_regression=LinearRegression()
linear_regression.fit(x_train,y_train)
model_pred_train=linear_regression.predict(x_train)
```

```python
train_error=mean_squared_error(y_train,model_pred_train)
```

```python
model_pred_test=linear_regression.predict(x_test)
test_error=mean_squared_error(y_test,model_pred_test)
```

```python
print(f'''

Train Error: {train_error}
Test Error:  {test_error}

''')
```

> - Here train error is higher that test error so this is underfitting problem
> - To fix that issue we will use regularization

# 4) Regularization

To fix underfitting we will
- Increasing the regularization parameter
- Use smaller sets of features
- Collect more training data

```python
from sklearn.linear_model import Ridge
```

```python
reg=Ridge(alpha=1e-09)
reg.fit(x_train,y_train)
reg_prediction_train=reg.predict(x_train)
reg_prediction_test=reg.predict(x_test)
```

```python
0.001**3
```

```python
train_error_reg=mean_squared_error(y_train,reg_prediction_train)
test_error_reg=mean_squared_error(y_test,reg_prediction_test)
```

```python
print(f'''

Train Error: {train_error_reg}
Test Error:  {test_error_reg}

''')
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,reg_prediction_train,c='r')
plt.title('Train Data Prediction')
plt.show()
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,reg_prediction_test,c='r')
plt.title('Test Data Prediction')

plt.show()
```
