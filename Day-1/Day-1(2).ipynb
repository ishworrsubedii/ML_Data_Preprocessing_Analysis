# Decision Tree

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('dataset.txt')
```

```python
dataset
```

- To implement Decision Tree we will have to use numberical data only
 
- we have to convert nationality to neumerial data like uk-1 usa-2 n-3

```python
# To convert numerical data we will use label encoder 
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
dataset.Nationality=label_encoder.fit_transform(dataset.Nationality)
dataset.Go=label_encoder.fit_transform(dataset.Go)
```

```python
a=dataset.Experience.values
b=np.mean(a)
```

```python
b
```

```python
fig, ax = plt.subplots(figsize=(10,8))
cor = dataset.corr()
sns.heatmap(cor, annot = True, cmap="YlGnBu")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import sys
feature_names=['Age','Experience','Rank','Nationality']
x=dataset[feature_names].values
y=dataset['Go'].values
d_tree=DecisionTreeClassifier()
d_tree=d_tree.fit(x,y)
predictionn=d_tree.predict(x)
tree.plot_tree(d_tree,feature_names=feature_names)

```

- Gini is use for splitting the data 
- Formula of Gini = 1 - (x/n)2 - (y/n)2
> where n=number of samples or data
      x=posative or 1
      y=negative or 0

```python
fig,ax=plt.subplots(1,4,figsize=(12,6),sharey=True)
for i in range(len(ax)):
    
    ax[i].scatter(x[:,i],y,label='Actual data',marker='^')
    ax[i].set_xlabel(feature_names[i])
    ax[i].scatter(x[:,i],predictionn,label='Model Prediction',c='r',marker='x')
ax[0].set_ylabel('GO')
plt.legend()
plt.show()
```
