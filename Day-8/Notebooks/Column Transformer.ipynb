```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(strategy='mean')
impute_age=si.fit_transform(x_train.Age.values.reshape(-1,1))
```

```python
x_train_pclass=x_train.Pclass.values.reshape(-1,1)
x_train_Age=impute_age
x_train_sex=one_hotencoded_sex
x_train_fare=x_train.Fare.values.reshape(-1,1)
X_trainn=np.concatenate((x_train_sex,x_train_pclass,x_train_Age,x_train_fare),axis=1)
# X_train = np.concatenate((x_train_pclass, x_train_sex, x_train_fare), axis=1)
```

## With Column Transformer

```python
from sklearn.compose import ColumnTransformer

transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(strategy='mean'),['Age']),
    
    ('tnf2',OneHotEncoder(sparse=False,drop='first'),['Sex'])
],remainder='passthrough')
```

```python
X_train
```

```python
X_train_column_transformer=transformer.fit_transform(X_train)
```
