# Univariate Analysis
> Here im going to use both iris and titanic datasets
-  It is used to analysis the value of one column each
## 1. Categorical Data

### i. CountPlot


```python
titanic.sample(10)
```

```python
fig,axes=plt.subplots(1,3,figsize=(12,6))
sns.countplot(data=titanic,x=titanic['Survived'],ax=axes[0])
sns.countplot(data=titanic,x=titanic['Embarked'],ax=axes[1])
sns.countplot(data=titanic,x=titanic['Sex'],ax=axes[2])

plt.plot()
```

### ii. Piechart
- It is used to represent the value in percentage form.

```python
labels=['male','female']
plt.pie(titanic.Sex.value_counts(),labels=labels,autopct='%2f')
plt.show()
```

```python
labels2=['Survived','Not Survived']
plt.pie(titanic.Survived.value_counts(),labels=labels2,autopct='%f')
plt.show()
```

## 2. Numerical Data

### i. Histogram

```python
fig,axes=plt.subplots(1,2,figsize=(12,6))
sns.histplot(x=titanic.Age,bins=20,ax=axes[0])
sns.histplot(x=titanic.Fare,bins=20,ax=axes[1])

plt.show()
```

```python

for i in range(3):
    max_fare=titanic.Fare.idxmax()    
    titanic=titanic.drop(max_fare)
```

```python
titanic.Fare.max()
```

- From here we can remove outlier and noise in data 

### ii. Distplot
- It is used to find skewness
- PDF(Probability Density Function)

- KDE(Kernal Desity Estimation)

```python
fig,axes=plt.subplots(1,2,figsize=(12,6))
sns.distplot(titanic.Fare,ax=axes[0])
sns.distplot(titanic.Age,ax=axes[1])
plt.show()
```

- The above plot shows that those who were paid more fair had more chances of surviving.

### iii. Boxplot

```python
fig,axes=plt.subplots(1,2,figsize=(12,6))
sns.boxplot(x=titanic.Age,ax=axes[0])
sns.boxplot(x=titanic.Fare,ax=axes[1])

plt.show()
```

- Here we can remove the those data which are outside the box and line

***Multivariate***

```python
sns.scatterplot(x=titanic.Fare,y=titanic.Age,hue=titanic.Survived)
plt.show()
```

### iv. Dist Plot(Numerical-Categorical)


```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
plt.plot()
```

### v. HeatMap(Categorical-Categorical)
- It will tell us closeness in two values


```python
titanic.sample(10)
```

```python
sns.heatmap(pd.crosstab(titanic.Pclass,titanic.Survived))
plt.show()
```

```python
sns.heatmap(titanic.pivot_table(values='Pclass',index='Parch',columns='Survived'))
```

### vi. Cluster Map(Categorical-Categorical)

```python
sns.clustermap(pd.crosstab(titanic.Pclass,titanic.Survived),) # Cross tab is used to see the relationship between two variables.
plt.show()
```
